## 一、前置作業

### （一）引入模組

```python
import matplotlib.pyplot as plt
```

### （二）讀取檔案

```python
with open('hw2_data.txt', 'r') as f:
  data = f.read().splitlines()
data
```

### （三）寫進hash裏面

```python
food_count = { }
for d in data:
  if d in food_count:
    food_count[d] += 1
  else:
    food_count[d] = 0
```

```python
food_count
```

##二、計算有多少個不重複的數字

```python
print(len(food_count.keys()))
```

## 三、計算每一個英文字出現次數為何

```python
for f in food_count.keys():
  print(f, '出現', food_count[f],'次')
```

## 四、畫圖

### （一）整理hash

```python
sorted(food_count)
```

```python
sorted_food_count = dict(sorted(food_count.items(), key=lambda item: item[1], reverse=True))
sorted_food_count
```

### （二）畫圖

```python
# 提取鍵和值
foods = list(sorted_food_count.keys())
counts = list(sorted_food_count.values())

# 畫條形圖
plt.bar(foods, counts, color='pink', edgecolor='black')

# 設定標題與標籤
plt.title('The time each food is counted', fontsize=16)
plt.xlabel('type of food', fontsize=12)
plt.ylabel('times', fontsize=12)

plt.xticks(rotation=45)  # 食物名稱旋轉以避免重疊
plt.show()
```
